# 3) Visualizations
1) Univariate
2) Bivariate
3) Multivariate

```python
#Univariate
df['Date'].plot()
plt.show()
df['Price'].plot()
```

```python
sns.distplot(df['Price'])
plt.show()
sns.distplot(df['number of bedrooms'])
plt.show()
```

```python
#bivariate
plt.scatter(df['number of bedrooms'],df.Price)
plt.show()
sns.lineplot(x=df['Date'],y=df.Price)
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
#bivariate-Joinplot
sns.jointplot(x='Date',y='Price',data=df)
plt.show()
sns.jointplot(x='number of bedrooms',y='Price',data=df)
```

```python
#Multivariate
sns.pairplot(df)#commented

#Output for paiplot has been attached as an immage in order to avoid file size exceeding 10MB
from matplotlib import image as img
image = img.imread("C:\\Users\\varun\\OneDrive\\Desktop\\pairplot op.png")
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```
